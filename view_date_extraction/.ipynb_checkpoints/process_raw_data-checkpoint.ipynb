TASK: model 1

TraningReadiness = Wellness 里的5个x variables → link to bigger model: 前一天的traningreadiness 和 当天的outcome bound

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import math
```

```python
standardized_wellness = pd.read_csv('./data/standardized_wellness.csv')
wellness = pd.read_csv('./data/wellness.csv', parse_dates=['Date'])
rpe = pd.read_csv('./data/rpe.csv', parse_dates=['Date'])
games = pd.read_csv('./data/games.csv', parse_dates=['Date'])
```

```python
standardized_wellness = pd.DataFrame(standardized_wellness)
wellness = pd.DataFrame(wellness)
rpe = pd.DataFrame(rpe)
games = pd.DataFrame(games)
```

```python
cols1 = ["Date","PlayerID"]
dat1 = wellness[cols1]

cols2 = ["TrainingReadiness"]
dat2 = wellness[cols2]

cols3 = ["Fatigue","Soreness", "Desire", "Irritability", "SleepQuality"]
dat3 = standardized_wellness[cols3]

final = pd.concat([dat1, dat3], axis=1)
model1 = pd.concat([final, dat2], axis=1)

model1['TrainingReadiness'] = model1['TrainingReadiness'].astype(str)

model1['TrainingReadiness'] = model1['TrainingReadiness'].str.rstrip('%').astype('float') /100.0

model1
```

TASK 3: teampoints/ totalpoints 某几个国家的 = Accurate…..+ training readiness

AcuteChronicRatio from rpe
training readiness from standardized wellness
points ratio from games

date playerid accurate fatigue 

```python
model2_after_categorical = pd.read_csv('./data/model2_after_categorical.csv',parse_dates=['Date'])
model2_after_categorical = pd.DataFrame(model2_after_categorical)
```

```python
training = rpe[["Date", "PlayerID", "AcuteChronicRatio"]]

readiness = model2_after_categorical[["Date", "PlayerID", "TrainingReadiness"]]

points = games[["Date", "Opponent", "TeamPoints", "TeamPointsAllowed"]]

dates = games[["Date"]]

training_avg = training.groupby('Date')['AcuteChronicRatio'].mean()
readiness_avg = readiness.groupby('Date')['TrainingReadiness'].mean()

training_avg = pd.DataFrame(training_avg)
training_avg["Date"] = training_avg.index.astype(str)
training_avg.index = range(len(training_avg))
training_avg
```

```python

readiness_avg = pd.DataFrame(readiness_avg)
readiness_avg["Date"] = readiness_avg.index.astype(str)
readiness_avg.index = range(len(readiness_avg))
readiness_avg

```

join two ratios

```python
# merge1 = readiness_avg.join(training_avg, on='Date')
# merge1

# merge1 = pd.concat([readiness_avg, training_avg], axis=1)

merge1 = readiness_avg.merge(training_avg, on="Date", how="inner")

merge1

```

```python
PointsRatio = [row.TeamPoints/(row.TeamPointsAllowed + row.TeamPoints) for index, row in points.iterrows()]

points["PointsRatio"] = PointsRatio
points = points[["Date", "Opponent", "PointsRatio"]]
points
```

```python
# answer = pd.concat([points, merge1], axis=1)

merge1["Date"] = merge1["Date"].astype(str)
points["Date"] = points["Date"].astype(str)

model3 = merge1.merge(points, on="Date", how="inner")

model3 = model3[["Date", "Opponent", 'TrainingReadiness', 'AcuteChronicRatio','PointsRatio']]

model3 = model3.sort_values('Opponent')

model3
```

```python
model1.to_csv('./data/model1.csv')

model2.to_csv('./data/model2.csv')

opponents = model3['Opponent'].unique().tolist()
opponents = pd.DataFrame(opponents)

opponents.columns = ['countries']
opponents


opponents.to_csv('./data/list_of_opponents.csv')
```

```python
model3["Accategorical"] = model3["AcuteChronicRatio"].apply(lambda x: "high" if x > 1.2 else "recover" if x < 0.8 else "normal")
model3["Trcategorical"] = model3["TrainingReadiness"].apply(lambda x: 1 if x > 0 else 0)

model3
model3.to_csv('./data/model3.csv')
```
